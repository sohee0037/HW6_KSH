```python
def main():
    import csv
    import matplotlib.pyplot as plt
    plt.rcParams["font.family"] = "NanumGothic"
    
    f = open('q4.csv', 'r', encoding='UTF-8')
    data = csv.reader(f)
    header = next(data)
    num_data = next(data)[0]
    on_people = []
    off_people = []
    total = []
    station_ls = []
    for row in data:
        on_people.append(int(row[10]) + int(row[12]))
        off_people.append(int(row[11]) + int(row[13]))
        total.append(int(row[10]) + int(row[11]) + int(row[12]) + int(row[13]))
        station_ls.append(row[3])
    f.close()
    
    sorted_on_people=sorted(on_people, reverse=True)
    sorted_off_people=sorted(off_people, reverse=True)
    sorted_total=sorted(total, reverse=True)
    
    on_station = []
    for i in sorted_on_people[:30]:
        index = on_people.index(i)
        station = station_ls[index]
        on_station.append(station)
    off_station = []
    for i in sorted_off_people[:30]:
        index = off_people.index(i)
        station = station_ls[index]
        off_station.append(station)
    tot_station = []
    for i in sorted_total[:30]:
        index = total.index(i)
        station = station_ls[index]
        tot_station.append(station)
    
    x = on_station
    y = sorted_on_people[:30]
    plt.xticks(rotation=90)
    plt.bar(x, y)
    plt.title(" ")
    plt.xlabel(" ")
    plt.ylabel(" ")
    plt.show()
    
    x = off_station
    y = sorted_off_people[:30]
    plt.xticks(rotation=90)
    plt.bar(x, y)
    plt.title(" ")
    plt.xlabel(" ")
    plt.ylabel(" ")
    plt.show()
    
    x = tot_station
    y = sorted_total[:30]
    plt.xticks(rotation=90)
    plt.bar(x, y)
    plt.title(" ")
    plt.xlabel(" ")
    plt.ylabel(" ")
    plt.show()

if __name__=='__main__':
    main()
```
