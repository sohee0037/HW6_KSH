```python
import csv
import matplotlib.pyplot as plt

def main():
    f = open('2008.csv', 'r', encoding='ANSI')
    data = csv.reader(f)
    header = next(data)
    men = []
    women = []
    for row in data:
        men.append(int(row[-3]))
        women.append(int(row[-2]))
    f.close()
    f = open('2011.csv', 'r', encoding='ANSI')
    data = csv.reader(f)
    header = next(data)
    for row in data:
        men.append(int(row[-3]))
        women.append(int(row[-2]))
    f.close()
    f = open('2015.csv', 'r', encoding='ANSI')
    data = csv.reader(f)
    header = next(data)
    for row in data:
        men.append(int(row[-3]))
        women.append(int(row[-2]))
    f.close()
    f = open('2019.csv', 'r', encoding='ANSI')
    data = csv.reader(f)
    header = next(data)
    for row in data:
        men.append(int(row[-3]))
        women.append(int(row[-2]))
    f.close()
    f = open('2022.csv', 'r', encoding='ANSI')
    data = csv.reader(f)
    header = next(data)
    for row in data:
        men.append(int(row[-3]))
        women.append(int(row[-2]))
    f.close()
    def index(i):
        data = [men[i], women[i]]
        label = ['men', 'women']
        plt.figure(figsize=(3, 3))
        plt.pie(data,labels=label, autopct='%.3f%%')
        plt.legend()
        plt.show()
        if float(men[i])/women[i] < 1:
            print("해당년도 제주도에서 여성의 비율이 더 높다.\n")
        else:
            print("해당년도 제주도에서 여성의 비율이 더 높지 않으므로 귀류법에 의해 제주도에는 여성의 비율이 더 높다는 속설은 거짓이다.\n")
        print()
        
    print('===2008 data===')
    index(0)
    print('===2011 data===')
    index(1)
    print('===2015 data===')
    index(2)
    print('===2019 data===')
    index(3)
    print('===2022 data===')
    index(4)

if __name__=='__main__':
    main()
```
